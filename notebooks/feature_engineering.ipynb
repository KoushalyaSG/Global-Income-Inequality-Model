```python
import pandas as pd

stage2_path = r"C:\Users\admin\Desktop\INTERN MINI PROJECT 1\proccessed\preprocessed_dataset.csv"

df = pd.read_csv(stage2_path)

print("Stage 2 data loaded")
print("Shape:", df.shape)
df.head()
```

```python
df['Education_Efficiency'] = (
    df['Education_Spending_%GDP'] /
    (df['Gini_Index'] + 1e-6)
)

df[['Education_Spending_%GDP','Gini_Index','Education_Efficiency']].head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))
sns.histplot(df['Education_Efficiency'], bins=30, kde=True)
plt.title("Distribution of Education Efficiency")
plt.show()
```

```python
df['Edu_to_Unemp_Ratio'] = (
    df['Education_Spending_%GDP'] /
    (df['Unemployment_%'] + 1e-6)
)

df[['Education_Spending_%GDP','Unemployment_%','Edu_to_Unemp_Ratio']].head()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Edu_to_Unemp_Ratio'], bins=30, kde=True)
plt.title("Education to Unemployment Ratio")
plt.show()
```

```python
df['Economic_Stress_Index'] = (
    df['Gini_Index'] + df['Unemployment_%']
)

df[['Gini_Index','Unemployment_%','Economic_Stress_Index']].head()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Economic_Stress_Index'], bins=30, kde=True)
plt.title("Economic Stress Index Distribution")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), cmap='coolwarm', annot=False)
plt.title("Feature Correlation Heatmap")
plt.show()
```
