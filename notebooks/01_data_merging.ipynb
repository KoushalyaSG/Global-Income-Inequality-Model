# Data Merging Notebook
This notebook merges Gini Index, Education Spending, and Unemployment data from World Bank Excel files.

```python
import pandas as pd
```

### Load Raw Files
Loading `.xls` files from the `data/raw/` directory.

```python
# Load raw datasets (Excel format)
gini = pd.read_excel("../data/raw/gini_index.xls", skiprows=3)
education = pd.read_excel("../data/raw/eduacation spending.xls", skiprows=3)
unemployment = pd.read_excel("../data/raw/unemployment_data.xls", skiprows=3)

print("Datasets loaded successfully.")
```

### Filter Countries

```python
countries = ["India", "United States", "Brazil"]

gini = gini[gini["Country Name"].isin(countries)]
education = education[education["Country Name"].isin(countries)]
unemployment = unemployment[unemployment["Country Name"].isin(countries)]
```

### Select Years (2014–2023)

```python
years = [year for year in range(2014, 2024)]

gini = gini[["Country Name"] + years]
education = education[["Country Name"] + years]
unemployment = unemployment[["Country Name"] + years]
```

### Convert Wide to Long Format

```python
gini = gini.melt(id_vars="Country Name", var_name="Year", value_name="Gini_Index")
education = education.melt(id_vars="Country Name", var_name="Year", value_name="Education_Spending_%GDP")
unemployment = unemployment.melt(id_vars="Country Name", var_name="Year", value_name="Unemployment_%")
```

### Merge Datasets

```python
# Merge step-by-step
merged_df = pd.merge(gini, education, on=["Country Name", "Year"], how="inner")
merged_df = pd.merge(merged_df, unemployment, on=["Country Name", "Year"], how="inner")

merged_df.rename(columns={"Country Name": "Country"}, inplace=True)
merged_df["Year"] = merged_df["Year"].astype(int)

print("Datasets merged successfully.")
merged_df.head()
```

### Save Processed Dataset

```python
import os
os.makedirs("../data/processed", exist_ok=True)
merged_df.to_csv("../data/processed/merged_dataset_v1.csv", index=False)

print("Merged dataset saved to processed folder.")
```
